```python
import pandas as pd
```

### Import Data

```python
# Import data
df09 = pd.read_csv('2009.csv', low_memory=False)
df10 = pd.read_csv('2010.csv', low_memory=False)
df11 = pd.read_csv('2011.csv', low_memory=False)
df12 = pd.read_csv('2012.csv', low_memory=False)
df13 = pd.read_csv('2013.csv', low_memory=False)
df14 = pd.read_csv('2014.csv', low_memory=False)
df15 = pd.read_csv('2015.csv', low_memory=False)
df16 = pd.read_csv('2016.csv', low_memory=False)
df17 = pd.read_csv('2017.csv', low_memory=False)
df18 = pd.read_csv('2018.csv', low_memory=False)
```

```python
df = pd.concat([df09, df10, df11, df12, df13, df14, df15, df16, df17, df18])
df.head()
```

```python
df.shape[0] # number of data entries
```

### Clean Data

```python
# Select columns of interest
columns = ['FL_DATE', 'OP_CARRIER', 'ORIGIN', 'DEST', 'CRS_DEP_TIME', 'DEP_DELAY', 'CANCELLED']
dat = df[columns]
dat
```

```python
# Remove flights that were cancelled
dat = dat[dat.CANCELLED == 0].drop(['CANCELLED'], axis=1)
dat
```

```python
# Remove missing data
dat = dat.dropna()
dat
```

```python
# Confirm not missing any data
missing_dat = dat.isnull().sum(axis=0).reset_index()
missing_dat.columns = ['variable', 'missing values']
missing_dat['filling factor (%)']=(dat.shape[0]-missing_dat['missing values'])/dat.shape[0]*100
missing_dat.sort_values('filling factor (%)').reset_index(drop = True)
```

### Modify Abbreviations

```python
# Get unique values
airlines = dat.OP_CARRIER.unique().tolist()
origins = dat.ORIGIN.unique().tolist()
destinations = dat.DEST.unique().tolist()

airlines.sort()
origins.sort()
destinations.sort()
```

```python
origins==destinations
```

```python
len(origins)
```

```python
# airport "FLN" is only in destination and not origin
poss_airports = origins + ["FLN"] # add to list of airports
poss_airports.sort()

# confirm it worked
len(poss_airports), len(origins)
```

```python
poss_airports
```

```python
airlines
```

```python
# Replace airline IATA abbreviations with name
# https://www.iata.org/en/publications/directories/code-search/
# https://en.wikipedia.org/wiki/List_of_airlines_of_the_United_States#Mainline
# https://en.wikipedia.org/wiki/List_of_airline_codes

dat['OP_CARRIER'].replace({
    '9E':'Endeavor Air',
    'AA':'American Airlines',
    'AS':'Alaska Airlines',
    'B6':'JetBlue Airways',
    'CO':'Continental Airlines',
    'DL':'Delta Airlines',
    'EV':'ExpressJet Airlines',
    'F9':'Frontier Airlines',
    'FL':'AirTran',
    'G4':'Allegiant Air',
    'HA':'Hawaiian Airlines',
    'MQ':'Envoy Air',
    'NK':'Spirit Airlines',
    'NW':'Northwest Airlines',
    'OH':'PSA Arlines',
    'OO':'SkyWest Airlines',
    'UA':'United Airlines',
    'US':'US Airways',
    'VX':'Virgin America',
    'WN':'Southwest Airlines',
    'XE':'JSX',
    'YV':'Mesa Airlines',
    'YX':'Republic Airways'
}, inplace=True)

dat
```

```python
dat.OP_CARRIER.value_counts()
```

```python
# Get file with airport information
airports = pd.read_csv('airports.csv', low_memory=False)
```

```python
airports = airports.loc[airports['iata_code'].isin(poss_airports)]

cols = ['name','latitude_deg','longitude_deg','iata_code']
airports = airports[cols].sort_values(by=['iata_code']).reset_index().drop(['index'], axis=1)

airports = airports.rename(columns={'name':'NAME', 'latitude_deg':'LAT', 'longitude_deg':'LONG', 'iata_code':'IATA'})
airports
```

```python
# save airport data as .csv
airports.to_csv('airports_filtered.csv',index=False)
```

```python
# Change airports of "PFN" to "ECP"
dat['ORIGIN'].replace({'PFN':'ECP'}, inplace=True)
dat['DEST'].replace({'PFN':'ECP'}, inplace=True)
```

```python
# save flight data as .csv
dat.to_csv('flight_data.csv',index=False)
```

```python
# save as smaller files for uploading to GitHub
NUM_OF_LINES=1000000
filename = 'flight_data.csv'
with open(filename) as fin:
    fout = open("flights0.csv","w")
    for i,line in enumerate(fin):
      fout.write(line)
      if (i+1)%NUM_OF_LINES == 0:
        fout.close()
        fout = open("flights%d.csv"%(i/NUM_OF_LINES+1),"w")

    fout.close()
```
